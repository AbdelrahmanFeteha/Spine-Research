Data details:
- Pelvic Incidence: The angle ebtween the line perpendicular to the sacral plate and the line connecting the midpoint of the sacral plate to the axis of the femoral heads. Refelcts the pelvic structure and can influence posture of the spine (Naturally forward-tilted. 35 - 85, average 51.9)

- Pelvic Tilt: The angle between the vertical and the line connecting the midpoint of the sacral plate to the axis of the femolar heads. Measures the orientation of the pelvis and chnages with conditions like spondylolisthesis or abnomrmal lumbar spine posture, affecting spinal stability (The average ranges of anterior and posterior pelvic tilting are 13.0 ± 4.9°, and 8.9 ± 4.5°)

- Lumbar Lordosis Angle: Inward curve of the lumbar spine. A higher or lower than normal lumbar lordosis angle may be associated with spinal conditions as the lumbar curve's shape is altered in hernia and spondylolisthesis patients (Ideally 40 - 60)

- Sacral Slope: The angle between the horizontal and the sacral plate. Measures the inclination of the sacral plate and is directly related to the pelvic incidence.  Changes in sacral lope can affect spinal alignment, contributing to conditions like hernia and spondylolithesis (between 35 and 45)

- Pelvic Radius: A measurement of the radius of the pelvis, liekly referring to a structural parameter of the pelvic region. Could indicate structrual integrity or deformity in the pelvis which might influence suceptibility to spinal disorders

- Degree Spondylolithesis: The forward or bckward displacement of a vertebra in relation to the vertebra below it. Key indicator of spondylolithesis, as this condition involves vertebral slippage, which directly affects this measurement (Grade I spondylolisthesis is 1 to 25% slippage, grade II is up to 50% slippage, grade III is up to 75% slippage, and grade IV is 76-100% slippage)


Categories:
- Hernia: Often involves abnormal spinal curvature and alignment, so deviations in features like pelvic incidence, pelvic tilt, and lumbar lordosis angle may indicate this condition.

- Spondylolisthesis: Characterized by vertebral slippage, which is reflected in a high degree of spondylolisthesis. Other features like pelvic tilt and sacral slope can also be affected by the vertebral misalignment.

- Normal: The values of these features would typically fall within a standard range for a healthy spine without major deformations.


# Decision Tree Classifier




```python
from sklearn import tree
from sklearn.model_selection import train_test_split

y = df['class']
x = df.drop('class', axis = 1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth' : [3, 5, 10, None],
    'min_samples_split' : [2, 10, 20],
    'min_samples_leaf' : [1, 5, 10],
    'max_features' : ['auto', 'sqrt', 'log2']
}

grid_search = GridSearchCV(DecisionTreeClassifier(criterion = 'entropy', random_state=42, class_weight = 'balanced'), param_grid, cv=5, n_jobs=1)
grid_search.fit(x_train, y_train)

# Print best parameters
print(grid_search.best_params_)
best_tree_model = grid_search.best_estimator_
print(best_tree_model)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)

model = DecisionTreeClassifier(criterion='entropy', random_state=42, class_weight='balanced')
model.fit(x_train_resampled, y_train_resampled)
```

```python
import pydotplus
from IPython.display import Image

dot_data = tree.export_graphviz(model, feature_names=x.columns, class_names=['Normal','Hernia', 'Spondylolisthesis'], filled=True,
                                out_file=None) # generates a GraphViz representation of the decision tree
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
score_list = model.feature_importances_
score_df = pd.DataFrame({'Feature': x.columns, 'Score': score_list})
score_df.sort_values('Score', ascending=False)
```

```python
list_of_features = list(x.columns)
plt.figure(figsize = (8,6))
plt.barh(range(len(list_of_features)), model.feature_importances_)
plt.yticks(np.arange(len(list_of_features)), list_of_features)
plt.ylabel('Features')
plt.show()
```
